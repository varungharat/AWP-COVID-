```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid_data = pd.read_csv("covid_19_india.csv")

covid_data
```

```python
covid_data['State/UnionTerritory'] = covid_data['State/UnionTerritory'].replace('Bihar****','Bihar')
covid_data['State/UnionTerritory'] = covid_data['State/UnionTerritory'].replace('Telengana','Telangana')
```

```python
plt.figure(figsize=(12,8))

plt.show()
```

```python
covid_data['Date'] = pd.to_datetime(covid_data['Date'])
```

```python
covid_data['Year']= covid_data['Date'].dt.year
covid_data['Month']= covid_data['Date'].dt.month
covid_data['Day']= covid_data['Date'].dt.day
```

```python
data_2020 = covid_data[covid_data['Year'] == 2020]

data_2020
```

```python
fig  = plt.figure(figsize=(15,8))
sns.scatterplot(x='Cured',y='State/UnionTerritory',hue='Date',data=data_2020)
plt.tight_layout()
fig.show()
```

```python
data_2021 = covid_data[covid_data['Year'] == 2021]

data_2021
```

```python
sns.set_style(style='whitegrid')
sns.scatterplot(x='Cured',y='State/UnionTerritory',hue="Date",data=data_2021)
plt.tight_layout()
fig.show()
```
